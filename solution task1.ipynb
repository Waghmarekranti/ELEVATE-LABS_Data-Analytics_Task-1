```python
import pandas as pd
print(df.columns)

# Step 1: Load the dataset
df = pd.read_csv("F:\OneDrive\Desktop\Task 1\Customer Data.csv")  # <-- Make sure this file exists in your working directory

# Step 2: Calculate IQR for "Income"
Q1 = df["year_birth"].quantile(0.25)
Q3 = df["year_birth"].quantile(0.75)
IQR = Q3 - Q1

# Step 3: Define bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Step 4: Replace outliers with median
df.loc[((df["year_birth"] < lower_bound) | (df["year_birth"] > upper_bound)), "year_birth"] = df["year_birth"].median()
```
